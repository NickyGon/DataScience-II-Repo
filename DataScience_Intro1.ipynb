```python
Dagrupados3=D.groupby(['pclass','survived','sex'])
Dagrupados3.aggregate(
    {
        'age':np.mean,
        'fare':np.max
    }
)
```

```python
Dagrupados4=D.groupby(['pclass','sex'])
Dagrupados4.aggregate(
{
    'survived':np.mean,
    'age':[np.min,np.mean,np.max],
    'parch':np.mean
})
```

```python
Dagrupados4.aggregate(
{
    'survived':np.mean,
    'age':[lambda x: np.max(x)-np.min(x),np.mean],
    'parch':np.mean
})
```

```python
def Rango(x):
    return np.max(x)-np.min(x)
```

```python
Dagrupados4.aggregate(
{
    'survived':np.mean,
    'age':[Rango,np.mean],
    'parch':np.mean
})
```
