```python
def DummyColumn(Dataset,Columna):
    temp=pd.get_dummies(Dataset[Columna],prefix=Columna)
    n=temp.shape[1]
    columnasDummies=temp.columns
    temp=temp.drop(columnasDummies[n-1],axis=1)
    D=pd.concat([Dataset,temp],axis=1)
    D=D.drop(Columna,axis=1)
    return D
```

Creación de 1000 estudiantes de la Universidad entre 18 a 23 años, de carreras aleatorias correspondientes a la FIA (Facultad de Ingenierias y Arquitectura), con notas de una media de 97 y de desviación estandar de 5 puntos, y una indicación sobre su graduación por excelencia (igual o mayor a 90.00)

```python
Edad=[]
Carrera=[]
Nota=[]
Graduacion=[]
Codigo=[]
Genero=[]

carrerasFIA=["Ingeniería del Medio Ambiente y Energías Alternativas", "Arquitectura", 
             "Ingeniería de la Producción","Ingeniería de Sistemas Computacionales","Ingeniería de Petroleo y Gas Natural",
             "Ingeniería Electromecánica","Ingeniería Civil","Ingeniería de Sistemas Electrónicos y Telecomunicación",
             "Ingeniería Industrial y de Sistemas", "Departamento de Ciencias Exactas"]
carrerasFIA.sort()

Edad=np.random.randint(low=18,high=24,size=1000)
Carrera=np.random.choice(carrerasFIA,size=1000)
Nota=np.round(np.random.normal(loc=97,scale=5,size=1000),2)
Codigo=list(range(50000,51000))
Codigo2=np.random.randint(low=50000,high=51000,size=1000)
Genero=np.random.choice(['Masculino','Femenino'],size=1000)

for i in range(1000):
    if Nota[i]<90.00:
        Graduacion.append("Por Trabajo final")
    else:
        if Nota[i]>100.00:
            Nota[i]=100.00
        Graduacion.append("Por Excelencia")
```

```python
def Solucion1(dataset):
    d=dataset.dropna(subset=['sulfate','nitrate'],how='all')
    d.index=range(len(d))
    d.loc[:,'sulfate']=d.sulfate.fillna(0)
    d.loc[:,'nitrate']=d.nitrate.fillna(0)
    return d
```

```python
D1s=Solucion1(Dd1)
D1s
```

```python
D1s[D1s['sulfate']==0]
```

```python
def Solucion1_2(dataset):
    d=dataset.dropna(subset=['sulfate','nitrate'],how='all')
    d.index=range(len(d))
    d=d.fillna(0)
    return d
```

```python
D2s=Solucion1_2(Dd1)
D2s
```

```python
D2s[D2s['sulfate']==0]
```

```python
def Solucion2(d):
    d=d.fillna(0)
    return d
```

```python
def LoadAllDist(option):
    archstr=''
    databases=[]
    for i in range(1,333,1):
        route='distributed-data/'
        ind=str(i)
        if i<10:
            archstr='00'+ind
        elif i<100:
            archstr='0'+ind
        else:
            archstr=ind
        route+=archstr+'.csv'
        datatemp=pd.read_csv(route)
        datab=Solucion2(datatemp)
        if (option==1):
            databases.append(datab)
        if (option==2):
            if i==1:
                databases=datab
            else:
                databases=pd.concat([databases,datab])
                databases.index=range(len(databases))
    return databases
```
